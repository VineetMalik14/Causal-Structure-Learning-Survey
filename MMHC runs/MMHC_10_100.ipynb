```python
import pandas as pd
import numpy as np
from pgmpy.estimators import MmhcEstimator
data = pd.read_csv('causal_10_100.csv')
data['sum'] = data.sum(axis=1)
est = MmhcEstimator(data)
model = est.estimate()
print(model.edges())
```

```python
from pgmpy.estimators import PC
est = PC(data)
skel, sep_sets = est.build_skeleton()
print(skel.edges())
print(sep_sets)
```
